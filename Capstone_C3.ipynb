```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
car_sales = pd.read_csv('./Car_sales_transactions.csv')
```

```python
car_sales
```

```python
car_sales.replace({'Transmission': {'Manual': 0, 'Automatic': 1}}, inplace=True)
```

```python
car_sales.Transmission
```

```python
car_sales.shape
```

```python
round(car_sales.describe(),2).T
```

```python
car_sales.info()
```

```python
car_sales.columns
```

```python
car_sales['Brand'] = car_sales['Name'].apply(lambda x: ' '.join(x.split()[:1]))
car_sales['Model'] = car_sales['Name'].apply(lambda x: ' '.join(x.split()[1:2]))
car_sales['Variant'] = car_sales['Name'].apply(lambda x: ' '.join(x.split()[2:]))
```

```python
car_sales
```

Which brand sells the most cars?

```python
car_sales.Brand.value_counts()
```

What are the average selling prices of the brands?

```python
sns.boxenplot(data=car_sales, x=car_sales.Selling_Price, y=car_sales.Brand)
```

How does the distance travelled by the car affect the selling price of the car?

```python
sns.scatterplot(data=car_sales, x=car_sales.km_Driven, y=car_sales.Selling_Price)
```

```python
car_sales.corr()
```

```python
sns.heatmap(car_sales.corr(), annot=True)
```

What is the average selling price compared to the number of seats of the car?

```python
sns.boxenplot(data=car_sales, y=car_sales.Selling_Price, x=car_sales.Seats)
```

```python
car_sales.Seats.value_counts()
```

How has the average price of cars fluctuated over the years?

```python
sns.lineplot(data=car_sales, x=car_sales.Year, y=car_sales.Selling_Price)
```

```python
car_sales.Fuel.value_counts()
```

```python
sns.countplot(data=car_sales, x=car_sales.Fuel)
```

What is the average Selling Price of cars having different fuel types?

```python
car_sales.groupby(by=car_sales.Fuel).mean().Selling_Price
```

```python
sns.countplot(data = car_sales, x=car_sales.Transmission)
```

```python
sns.countplot(data=car_sales, x=car_sales.Owner)
```

How does the maximum power of the car affect the selling price of the car?

```python
sns.lineplot(data=car_sales, x=car_sales.Max_Power, y=car_sales.Selling_Price )
```
